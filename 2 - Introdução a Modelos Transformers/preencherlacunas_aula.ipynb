```python
# Importa o pacote transformers e a função pipeline.
import transformers
from transformers import pipeline
```

```python
# Cria um pipeline de preenchimento de lacunas com modelo em português.
mascarar = pipeline("fill-mask", model="neuralmind/bert-base-portuguese-cased")
```

```python
# Executa predição para uma frase com máscara e imprime as sugestões.
texto = mascarar("Existe uma chance do copo cair no [MASK]")
for x in range(len(texto)):
  print(texto[x])
```

```python
# Executa predição para duas frases com máscara e mostra os resultados.
texto = mascarar(["Existe uma chance do copo cair no [MASK]","Brasília é a capital do [MASK]"])
for x in range(len(texto)):
  print(texto[x])
```
